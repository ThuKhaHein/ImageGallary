```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart (Myanmar trade data in USD million)
years = ['2023-2024', '2024-2025']
exports = [4644.75, 4412.41]  # Exports
imports = [5454.34, 3949.09]  # Imports

# Set up bar positions
bar_width = 0.35
index = np.arange(len(years))

# Create the bar chart
plt.bar(index, exports, bar_width, label='Exports', color='blue')
plt.bar(index + bar_width, imports, bar_width, label='Imports', color='orange')

# Add labels and title
plt.xlabel('Fiscal Year')
plt.ylabel('Value (USD Million)')
plt.title('Myanmar Exports and Imports')
plt.xticks(index + bar_width / 2, years)
plt.legend()

# Save the chart as a PNG file
plt.savefig('trade_chart.png')

# The chart will also be displayed below the cell automatically
```
